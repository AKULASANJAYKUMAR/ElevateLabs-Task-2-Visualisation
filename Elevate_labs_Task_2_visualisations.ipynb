```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("classic")  # simple clean theme
```

```python
df = pd.read_csv("superstore_dataset.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()

```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

plt.figure(figsize=(10,5))
monthly.plot()
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby("Sub-Category")["Profit"].sum().sort_values().plot(kind="barh")
plt.title("Profit by Sub-Category")
plt.xlabel("Profit")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(6,6))
region_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.show()
```
